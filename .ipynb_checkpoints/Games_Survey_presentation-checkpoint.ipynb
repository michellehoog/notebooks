## We surveyed 12 Umuzi recruits to learn about their gaming habits

```python
import pandas
import matplotlib.pyplot as plt
%matplotlib inline

survey = pandas.read_csv("Worksheet_Michelle.xlsx - Worksheet.csv")
```

## 62% of recruits use their cell phones to play games

```python
devices = survey.groupby("What device(s) do you use to play games?").size().reset_index(name = "count")
labels = ["Cell", "Cell & Desktop", "Desktop", "Game console", "Laptop"]

plt.bar(devices['What device(s) do you use to play games?'], devices['count'], tick_label=labels, align='edge')
plt.title("Devices Used To Play Games", fontsize = 16)
plt.xticks(rotation = 60)
```

## Sport was the most popular genre.
33% of recruits play multiple genres.

```python
genres = survey.groupby('What genres do you play?').size() .reset_index(name = "count")
explode = (0, 0, 0, 0, 0.1)  # explode sports
 
plt.pie(genres['count'], labels=genres['What genres do you play?'], autopct='%1.0f%%', startangle=90, explode = explode)
plt.title("Preferred Game Genres", fontsize = 16)
```

## Most recruits play for 1-2 hours per day

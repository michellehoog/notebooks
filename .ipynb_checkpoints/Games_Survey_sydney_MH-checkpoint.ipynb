# Game Survey Results


We recently conducted a survey, and the object of this notebook is to use Python's data visualisation packages to visualize the data.

First, let's load the the data and display it in a table.

```python
import pandas

survey = pandas.read_csv("Worksheet_Michelle.xlsx - Worksheet.csv")
survey
```

Now, let's begin to break it down so as to better process the feedback we got.

First, let's make a pie chart of the devices used to play games.

```python
import matplotlib.pyplot as plt
labels = 'Cellphone', 'Laptop', 'Desktop', 'Game Console'
count = [8, 1, 2, 2]
colors = ['LightSalmon', 'PaleGoldenrod', 'Teal', 'SlateBlue']

plt.pie(count, labels=labels, autopct='%1.0f%%', startangle=90, shadow=True)
plt.title('Devices Used To Play Games', fontsize = 16)
plt.show();
```

```python
labels = 'Sports', 'Action', 'Maze & Puzzles', 'Adventure', 'Multiple Genres'
count = [4, 2, 1, 1, 4]
colors = ['LightSalmon', 'PaleGoldenrod', 'Teal', 'SlateBlue', 'DarkSlateGray']

plt.pie(count, labels=labels, autopct='%1.0f%%', startangle=180, shadow=True)
plt.title("Preferred Game Genres")
plt.show()
```

If we were now to create the same graph, except this time automatically creating a series from the dataframe using the dataframe's "groupby()" method, we have the following result

```python
genres = survey.groupby('What genres do you play?').size()

labels = ["Action", "Adventure", "Maze & Puzzles", "Multiple Genres", "Sports"]

plt.pie(genres, labels=labels, autopct='%1.0f%%', startangle=180, shadow=True)
plt.title("Preferred Game Genres")
plt.show()
```

```python
#wrong way to plot
plt.hist(survey['Hours per day'])
plt.xlabel('Time playing (Hours)')
plt.ylabel("Frequency")
plt.title("Time spent playing computer games")
```

```python
#correct way
time = survey.groupby('Hours per day').size().reset_index(name='count') .rename(columns={'Col1':'Time'})
time

plt.bar(time['Hours per day'], time['count'])
plt.title('Hours spent playing computer games')
plt.ylabel("Frequency")
plt.show();
```

# Importing libraries
## what are libraries
A package is a collection of python modules under a common namespace. In practice one is created by placing multiple python modules in a directory with a special __init__.py module (file). A module is a single file of python code that is meant to be imported
### Numpy & MatPlotLib

NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

MatPlotLib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general

```python
# Definition of radius
r = 0.43

# Import the math package
import math

# Calculate C
C = 0
C = 2*r*math.pi
# Calculate A
A = 0
A = math.pi*r**2
# Build printout
print("Circumference: " + str(C))
print("Area: " + str(A))
```

# Numpy

```python
# Importimg numpy as np
import numpy as np

# changing age and height into arrays
np_age = np.array(age)
np_height = np.array(height)

# printing out np_age type
print(type(np_age))

# printing out np_height shape
print(np_height.shape)

#lets try printing the average height of the people's heights.
print(np.mean(np_height[:]))
print(np.median(np_height[:]))
```

```python
import matplotlib.pyplot as mplot
ageplot = np_age[:,:1]
ageplot

xlab = 'student names'
ylab = 'height [in meters]'
title= 'Age of students per by years'

mplot.xlabel('student names')
mplot.ylabel('height [in meters]')
mplot.title('Age of students per by years')

mplot.scatter(ageplot,height)
mplot.show
```
